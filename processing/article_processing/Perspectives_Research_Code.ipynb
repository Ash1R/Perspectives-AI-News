```python
#console ver
apik = ""

import requests
import os
import traceback

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from tqdm import tqdm
import re
import bs4

import firebase_admin
from firebase_admin import credentials
from firebase_admin import db
from firebase_admin import firestore

import gensim
import gensim.downloader as api
from bs4 import BeautifulSoup
from bs4.element import Comment
import functions_framework
import openai
import ot
import time

from datetime import date

@functions_framework.http
def hello_http(request):
  """HTTP Cloud Function.
  Args:
  request (flask.Request): The request object.
  <https://flask.palletsprojects.com/en/1.1.x/api/#incoming-request-data>
  Returns:
  The response text, or any set of values that can be turned into a
  Response object using `make_response`
  <https://flask.palletsprojects.com/en/1.1.x/api/#flask.make_response>.
  """
  request_json = request


  db = firestore.client()
  country = request_json["country"]
  date= request_json["date"]
  yesterday = request_json["yesterday"]
  categories = ["general", "business", "technology", "science", "sports", "entertainment", "health"]
  id = 1
  for cat in categories:
    print("https://us-central1-perspectives-27600.cloudfunctions.net/top-headlines1?country=us&from="+yesterday+"&to=" + date +"&category=" + cat + "&type=top-headlines&replacer=placeholder&apik=" + apik)
    data = requests.get("https://us-central1-perspectives-27600.cloudfunctions.net/top-headlines1?country=us&from="+yesterday+"&to=" + date +"&category=" + cat + "&type=top-headlines&replacer=placeholder&apik=" + apik).json()
    comb_text = {}
    print(data)
    for i in data["articles"][:4]:
      in_text = i["title"]
      if i["description"]:
        in_text += i["description"]
      if i["content"]:
        in_text += i["content"]
      comb_text[in_text] = i
   # headers = {"Authorization": f"Bearer {openkey}"}
    keywords = {}
    print(comb_text)
    for i in comb_text:
      storieslst =[]
      if i:
        try:
          print("before sending the excerpt to openai")
          convo = model.start_chat(history=[])
          convo.send_message("Read this excerpt and answer the question after the & signs:\n" +i[:100] + "\n&&&&&&&&&&\n What is the topic of this article, defined in 2 words or less, that I can use to find other articles on this topic, in the excerpt? Answer with JUST the 2 or less words.")
          print(convo.last.text)

          print("response from openai")
          #print(response)
          keywords[convo.last.text] = comb_text[i]
          time.sleep(3)
        except Exception:
          traceback.print_exc()
          print("keyword failed")
    #    google_news = api.load("glove-wiki-gigaword-100")

    google_news = api.load("glove-wiki-gigaword-50")
    for i in keywords:
      print(keywords)
      print(len(keywords))
      print("flag")
      biases = {}

      print(i)
      replacer = i
      topicart = requests.get("https://us-central1-perspectives-27600.cloudfunctions.net/top-headlines1?country=us&from=" + yesterday + "&to=" + date+ "&category=" + cat + "&type=everything&replacer=" + replacer+ "&apik=" + apik)
      topicson = topicart.json()
      for art in topicson["articles"][:4]:
        biases[str(google_news.wmdistance(extract(art["url"]), doc1))] = art
      biases[str(google_news.wmdistance(extract(keywords[i]["url"]), doc1))] = keywords[i]
      biases["date"] = str(date)
      biases["cat"] = cat
      biases["keyword"] = i
      biases["country"] = country
      doc_ref = db.collection("topics").document(str(id) + "-" + str(yesterday) + "-" + cat + "-" + i)
      print(biases)
      print("IT SHOULD SET HERE")
      doc_ref.set(biases)
      id+=1
  #print(storieslst)
  return storieslst


def extract(url):
  # Specify the URL of the news webpage


  # Send a GET request to the webpage
  print(url)
  try:
    response = requests.get(url)
  except:
    print("whoops")
    return "inf"


  # Create a BeautifulSoup object to parse the HTML content
  soup = BeautifulSoup(response.content, "html.parser")


  # Function to check if an element is a tag containing visible text
  def tag_visible(element):
    if element.parent.name in ["style", "script", "head", "title", "meta", "[document]"]:
      return False
    if isinstance(element, Comment):
      return False
    return True

  # Find all visible text elements on the webpage
  text_elements = soup.findAll(text=True)
  visible_texts = filter(tag_visible, text_elements)


  # Filter the visible text elements to get those longer than 100 characters
  long_texts = [re.sub('\s+',' ',text.strip()) for text in visible_texts if (len(re.sub('\s+',' ',text.strip())) > 100 and "newsletter" not in (text.strip().lower()) and "notification" not in text.strip().lower())]
  # Print the long texts
  return preprocessing((" ".join(long_texts)).split(" "))

def preprocessing(titles_array):
  final = []
  for title in tqdm(titles_array):
    processed = re.sub('[^a-zA-Z ]', '', title)

    words = processed.lower().split()
    stop_words = set(stopwords.words('english'))

    # keep words that have length of more than 1 (e.g. gb, bb), remove those with length 1.
    final.append(' '.join([word for word in words if (len(word) > 1 and word not in stop_words)]))
  return [i for i in final if i!= ""]


# Returns the current local date
from datetime import timedelta

today=date.today()


hello_http({"country": "us", "date": str(today), "yesterday": (str(today  - timedelta(days=1)))})



```

```python
#console ver
apik = ""

import requests
import os

import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from tqdm import tqdm
import re
import bs4

import firebase_admin
from firebase_admin import credentials
from firebase_admin import db
from firebase_admin import firestore

import gensim
import gensim.downloader as api
from bs4 import BeautifulSoup
from bs4.element import Comment
import functions_framework
import openai
import ot

from datetime import date

@functions_framework.http
def hello_http(request):
  """HTTP Cloud Function.
  Args:
  request (flask.Request): The request object.
  <https://flask.palletsprojects.com/en/1.1.x/api/#incoming-request-data>
  Returns:
  The response text, or any set of values that can be turned into a
  Response object using `make_response`
  <https://flask.palletsprojects.com/en/1.1.x/api/#flask.make_response>.
  """
  request_json = request


  db = firestore.client()
  country = request_json["country"]
  date= request_json["date"]
  yesterday = request_json["yesterday"]
  categories = ["general", "business", "technology", "science", "sports", "entertainment", "health"]
  id = 1
  for cat in categories:
    print("https://us-central1-perspectives-27600.cloudfunctions.net/top-headlines1?country=fr&from="+yesterday+"&to=" + date +"&category=" + cat + "&type=top-headlines&replacer=placeholder&apik=" + apik)
    data = requests.get("https://us-central1-perspectives-27600.cloudfunctions.net/top-headlines1?country=" + country+ "&from="+yesterday+"&to=" + date +"&category=" + cat + "&type=top-headlines&replacer=placeholder&apik=" + apik).json()
    comb_text = {}
    print(data)
    for i in data["articles"][:4]:
      in_text = i["title"]
      if i["description"]:
        in_text += i["description"]
      if i["content"]:
        in_text += i["content"]
      comb_text[in_text] = i
    openkey = ""
    headers = {"Authorization": f"Bearer {openkey}"}
    keywords = {}
    print(comb_text)
    for i in comb_text:
      storieslst =[]
      if i:
        data = {"model": "gpt-3.5-turbo", "messages": [{"role": "user", "content": "Read this excerpt and answer the question after the & signs:\n" + i + "\n&&&&&&&&&&\n What is the topic of this article, defined in 2 words or less, that I can use to find other articles on this topic, in the excerpt? Answer with JUST the 2 or less words."}], "temperature": 0.7}
        response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=data).json()
        try:
          print()
          keywords[response["choices"][0]["message"]["content"]] = comb_text[i]
        except:
          print("keyword failed")
    #    google_news = api.load("glove-wiki-gigaword-100")

    google_news = api.load("glove-wiki-gigaword-50")
    for i in keywords:
      print(keywords)
      print(len(keywords))
      print("flag")
      biases = {}

      print(i)
      replacer = i
      topicart = requests.get("https://us-central1-perspectives-27600.cloudfunctions.net/top-headlines1?country=us&from=" + yesterday + "&to=" + date+ "&category=" + cat + "&type=everything&replacer=" + replacer+ "&apik=" + apik)
      topicson = topicart.json()
      for art in topicson["articles"][:4]:
        biases[str(wv_from_bin.wmdistance(extract(art["url"]), doc1))] = art
      biases[str(wv_from_bin.wmdistance(extract(keywords[i]["url"]), doc1))] = keywords[i]
      biases["date"] = str(date)
      biases["cat"] = cat
      biases["keyword"] = i
      biases["country"] = country
      doc_ref = db.collection("topics").document(str(id) + "-" + str(yesterday) + "-" + cat + "-" + i+ "-" + country)
      print(biases)
      doc_ref.set(biases)
      id+=1
  #print(storieslst)
  return storieslst


def extract(url):
  # Specify the URL of the news webpage


  # Send a GET request to the webpage
  print(url)
  try:
    response = requests.get(url)
  except:
    print("whoops")
    return "inf"


  # Create a BeautifulSoup object to parse the HTML content
  soup = BeautifulSoup(response.content, "html.parser")


  # Function to check if an element is a tag containing visible text
  def tag_visible(element):
    if element.parent.name in ["style", "script", "head", "title", "meta", "[document]"]:
      return False
    if isinstance(element, Comment):
      return False
    return True

  # Find all visible text elements on the webpage
  text_elements = soup.findAll(text=True)
  visible_texts = filter(tag_visible, text_elements)


  # Filter the visible text elements to get those longer than 100 characters
  long_texts = [re.sub('\s+',' ',text.strip()) for text in visible_texts if (len(re.sub('\s+',' ',text.strip())) > 100 and "newsletter" not in (text.strip().lower()) and "notification" not in text.strip().lower())]
  # Print the long texts
  return preprocessing((" ".join(long_texts)).split(" "))

def preprocessing(titles_array):
  final = []
  for title in tqdm(titles_array):
    processed = re.sub('[^a-zA-Z ]', '', title)

    words = processed.lower().split()
    stop_words = set(stopwords.words('french'))

    # keep words that have length of more than 1 (e.g. gb, bb), remove those with length 1.
    final.append(' '.join([word for word in words if (len(word) > 1 and word not in stop_words)]))
  return [i for i in final if i!= ""]


# Returns the current local date
from datetime import timedelta

today=date.today()


hello_http({"country": "fr", "date": str(today), "yesterday": str(today - timedelta(days=1))})



```
